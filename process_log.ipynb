# Activity Monitor
Tracking linux box usage over time.

```python
import plotly.express as px
import datetime
from IPython.display import display, Markdown
from idle_log_processing import process_idle_log
```

```python
activity_info = process_idle_log('/export/bulk/tmp/idle.log')
```

## Today's Activity:

```python
lines = activity_info.report_str.split('\n')
for line in lines[1:3]:
    display(Markdown(line))
display(Markdown('\n'.join(lines[3:])))
```

```python
fig = px.line(activity_info.timeseries_today, x='timestamp', y='remaining_time', range_x=[activity_info.midnight + datetime.timedelta(hours=8),activity_info.midnight + datetime.timedelta(hours=23)], range_y=[0,8])
fig.update_layout(width=900, height=700)
fig.show()
```

# Activity Data for All Time

```python
#plot activity by day
fig = px.scatter(
    activity_info.activity_timeseries[activity_info.activity_timeseries.is_active],
    x='date', y='time_of_day', color='is_active'
)
fig.update_layout(
    title="Activity times by day",
    yaxis_tickformat = '%H:%M',
    yaxis = dict(autorange = 'reversed'),
    width=900, height=700
)

fig.show()
```

```python
#plot total activity by day
fig = px.bar(activity_info.daily_activity, x=activity_info.daily_activity.index, y='active_hours')
fig.update_layout(
    title="Daily activity totals",
    width=900, height=700
)
fig.show()
```

```python
#plot total activity by week
fig = px.bar(activity_info.weekly_activity, x=activity_info.weekly_activity.index.to_timestamp(), y='active_hours')
fig.update_layout(
    title="Weekly activity totals",
    width=900, height=700,
)
fig.show()
```
